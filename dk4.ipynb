```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating the DataFrame
data = {
    'Visitor_ID': [f'V{1000 + i}' for i in range(20)],
    'Name': ['ABHISHEK SANDEEP ZADE', 'ARNAV AJAY DESHPANDE', 'ASHWINI LALCHAND MUNDAWARE',
             'GAYATRI SURESH GAIKWAD', 'HARSHADA GANESH CHAUDHARI', 'VAIBHAVI HARISHWAR PATIL', 
             'VISHAKHA PUNDLIK JADHAV', 'YASH BHARAT SOLUNKE', 'VIVEK SANTOSH KHANDWE', 'Tanushree Chhanwal', 
             'Shruti Jaiswal', 'Shriyash Sulakhe', 'YASH BHARAT SOLUNKE', 'ARNAV AJAY DESHPANDE', 
             'RUTUJA SANTOSH THOTE', 'ROHIT DILIP BILWAL', 'RITESH SHIVAJI BAIRAGI', 'VISHAKHA PUNDLIK JADHAV', 
             'ROHIT DILIP BILWAL', 'RITESH SHIVAJI BAIRAGI'],
    'City': ['Delhi', 'Aurangabad', 'Mumbai', 'Mumbai', 'New Delhi', 'New Delhi', 'Chennai', 'Chennai', 'Bangalore',
             'Bangalore', 'Pune', 'Mumbai', 'Mumbai', 'Indore', 'Bangalore', 'Ujjain', 'Bangalore', 'Delhi', 'Pune', 'Bangalore'],
}

df = pd.DataFrame(data)

# Simulating Visit Dates (Randomized for 5 Days)
np.random.seed(42)
df['Visit_Date'] = pd.date_range(start='2024-06-01', periods=5).to_list() * 4

# Visualization 1: Number of Visitors Over Different Days (Line Plot)
visitors_per_day = df.groupby('Visit_Date').size()

plt.figure(figsize=(10, 5))
plt.plot(visitors_per_day.index, visitors_per_day.values, marker='o', linestyle='-', color='b', label="Visitors Count")
plt.xlabel("Date")
plt.ylabel("Number of Visitors")
plt.title("Number of Visitors Over Different Days")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.savefig("visitors_per_day.png", dpi=300)
plt.show()

# Visualization 2: Distribution of Participants by City (Bar Chart)
city_counts = df['City'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette='viridis')
plt.xlabel("City")
plt.ylabel("Number of Participants")
plt.title("Distribution of Participants by City")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(["Participants Count"])
plt.savefig("participants_by_city.png", dpi=300)
plt.show()
```
